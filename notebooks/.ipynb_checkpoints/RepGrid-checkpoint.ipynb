```python
#Find the valid data from the study

def get_index_study(csvRows):
    valid_index = []
    not_valid = ['sara', 'oliver', 'undefi']
    for i, vals in enumerate(csvRows):
        if [True for i in not_valid if i in vals[0]]:
            pass
        else:
            valid_index.append(i)
    return valid_index
valid_index = get_index_study(csvRows=csvRows)
csvRows_study = [csvRows[i] for i in valid_index]
```

```python
# 0- Not stressful 1 - Somewhat stressful 2 - Stressful 3 - Very stressful

def get_stress_score(csvRows):
    scores_ = []
    for ind in csvRows:
        score = 0
        for i, vals in enumerate(ind):
            if "stressful-situations-Punctual" in vals:
                start = i
                break
        stress_score = ind[start:start+16]
        aux = []
        for i, vals in enumerate(stress_score):
            if "not stressful" in vals and "somewhat stressful" not in vals:
                aux.append(0)
            elif "somewhat stressful" in vals:
                aux.append(1)
            elif "stressful" in vals:
                aux.append(2)
            elif "very stressful" in vals:
                aux.append(3)
        scores_.append(aux)
    return scores_
scores_ = get_stress_score(csvRows=csvRows_study)
```

## PSS, self-esteem and identity consistency

```python
def get_pss(csvRows):
    opposite = [3,4,6,7]
    scores_ = []
    for ind in csvRows:
        score = 0
        for i, vals in enumerate(ind):
            if "percieved-stress0" in vals:
                start = i
                break
        stress_score = ind[start:start+10]
        for i, vals in enumerate(stress_score):
            aux = 0
            if "never" in vals and "almost never" not in vals:
                aux=0
            elif "almost never" in vals:
                aux=1
            elif "sometimes" in vals:
                aux=2
            elif "fairly often" in vals:
                aux=3
            elif "very often" in vals:
                aux=4
            if i in opposite:
                score+=abs(aux-4)
            else: 
                if i not in opposite:
                    score+=aux
        scores_.append(score)
    return scores_
PSS = get_pss(csvRows=csvRows_study)
```

```python
import numpy as np

def get_repgrid(csvRows):
    grids = []
    for ind in csvRows:
        score = 0
        grid = []
        for i, vals in enumerate(ind):
            if 'rating_' in vals:
                start = i
                break
        ratings = ind[start:start+60]
        for i, vals in enumerate(ratings):
            score = vals.split(":")[1].replace('"',"")
            grid.append(int(score))
        grid = np.array_split(grid, 10)
        grids.append(grid)
    return grids
```

```python
# Calculate identity consistency as the mean variance over all situations  
def calc_ident_consistency(rep_grid):
    ignore = [0,1]
    mvar = []
    for i in range(len(rep_grid)): 
        variance = []
        #for j in range(len(rep_grid[i])): 
        for j in [x for x in range(len(rep_grid[i])) if x != 1]:
            var = np.var(rep_grid[i][j])
            variance.append(var)
        mvar.append(np.mean(variance))
    return mvar

consistency = calc_ident_consistency(rep_grid)
```

```python
def get_median(csvRows, startString):
    scores = []
    for ind in csvRows:
        for i, vals in enumerate(ind):
            if startString in vals:
                scores.append(vals.split(":")[1].replace('"',""))
    countSD = 0
    countD = 0
    countA = 0
    countSA = 0
    
    for val in scores:
        if val.lower() == 'strongly agree':
            countSA += 1
        elif val.lower() == 'agree':
            countA += 1
        elif val.lower() == 'disagree':
            countD += 1 
        elif val.lower() == 'strongly disagree':
            countSD += 1 
    vals = [1]* countSD + [2]* countD +[3]* countA +[4]* countSA
    m = np.median(vals)
    
    return m
m1 = get_median(csvRows_study,"repgrids_results1_question_2")
m2 = get_median(csvRows_study,"repgrids_results2_question_2")
m3 = get_median(csvRows_study,"repgrids_results3_question_2")
m4 = get_median(csvRows_study,"repgrids_results4_question_2")
idenm2 = get_median(csvRows_study,"identity_result_page_question_2")
mnew1 = get_median(csvRows_study,"repgrids_results1_question_1")
mnew2 = get_median(csvRows_study,"repgrids_results2_question_1")
mnew3 = get_median(csvRows_study,"repgrids_results3_question_1")
mnew4 = get_median(csvRows_study,"repgrids_results4_question_1")
idenm1 = get_median(csvRows_study,"identity_result_page_question_1")
```

```python
def get_mean(csvRows, startString):
    scores = []
    for ind in csvRows:
        for i, vals in enumerate(ind):
            if startString in vals:
                scores.append(vals.split(":")[1].replace('"',""))
    countSD = 0
    countD = 0
    countA = 0
    countSA = 0
    
    for val in scores:
        if val.lower() == 'strongly agree':
            countSA += 1
        elif val.lower() == 'agree':
            countA += 1
        elif val.lower() == 'disagree':
            countD += 1 
        elif val.lower() == 'strongly disagree':
            countSD += 1 
    vals = [1]* countSD + [2]* countD +[3]* countA +[4]* countSA
    m = np.mean(vals)
    
    return m
m1 = get_mean(csvRows_study,"repgrids_results1_question_2")
m2 = get_mean(csvRows_study,"repgrids_results2_question_2")
m3 = get_mean(csvRows_study,"repgrids_results3_question_2")
m4 = get_mean(csvRows_study,"repgrids_results4_question_2")
idenm2 = get_mean(csvRows_study,"identity_result_page_question_2")
mnew1 = get_mean(csvRows_study,"repgrids_results1_question_1")
mnew2 = get_mean(csvRows_study,"repgrids_results2_question_1")
mnew3 = get_mean(csvRows_study,"repgrids_results3_question_1")
mnew4 = get_mean(csvRows_study,"repgrids_results4_question_1")
idenm1 = get_mean(csvRows_study,"identity_result_page_question_1")
```

```python

def get_interesting(csvRows, startString):
    scores = []
    for ind in csvRows:
        for i, vals in enumerate(ind):
            if startString in vals:
                scores.append(vals.split(":")[1].replace('"',""))
    countSD = 0
    countD = 0
    countA = 0
    countSA = 0
    
    for val in scores:
        if val.lower() == 'strongly agree':
            countSA += 1
        elif val.lower() == 'agree':
            countA += 1
        elif val.lower() == 'disagree':
            countD += 1 
        elif val.lower() == 'strongly disagree':
            countSD += 1 
    total = countSA + countA + countD + countSD
    sa = countSA / total*100
    sd = countSD / total*100
    a = countA / total*100
    d = countD / total*100
    rep = []
    rep.append(round(sa,1))
    rep.append(round(a,1))
    rep.append(round(d,1))
    rep.append(round(sd,1))
    
    return rep

interesting1 = get_interesting(csvRows_study,"repgrids_results1_question_2")
interesting2 = get_interesting(csvRows_study,"repgrids_results2_question_2")
interesting3 = get_interesting(csvRows_study,"repgrids_results3_question_2")
interesting4 = get_interesting(csvRows_study,"repgrids_results4_question_2")
interestingIdent = get_interesting(csvRows_study,"identity_result_page_question_2")
learnt1 = get_interesting(csvRows_study,"repgrids_results1_question_1")
learnt2 = get_interesting(csvRows_study,"repgrids_results2_question_1")
learnt3 = get_interesting(csvRows_study,"repgrids_results3_question_1")
learnt4 = get_interesting(csvRows_study,"repgrids_results4_question_1")
learntIdent = get_interesting(csvRows_study,"identity_result_page_question_1")
```
